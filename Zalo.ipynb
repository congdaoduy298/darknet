```
import json 

def add_and_remove_bboxes(results, file_name):
  # Read file 
  with open(file_name, 'r') as f:
    data = f.read()
  # Get label id 
  for i in label:
    if i in file_name:
      label_id = label.index(i) + 1
      break 

  # Get all bboxes and save to json file with threshold = 0.25
  for line in data.split('\n'):
      arr = line.split()
      try:
        id, bbox = arr[0], arr[1:]
      except:
        print(arr)
      bbox = list(map(float, bbox))
      if bbox[0] > 0.25:
        width, height = bbox[3] - bbox[1], bbox[4] - bbox[2]
        results.append({"image_id": int(id),
                        "category_id": label_id,
                        "bbox": [round(bbox[1], 2), round(bbox[2], 2), 
                                 round(width, 2), round(height, 2)],
                        "score": bbox[0]})


label = ['Cam nguoc chieu', 'Cam dung va do', 'Cam re', 'Gioi han toc do', 
         'Cam con lai', 'Nguy hiem', 'Hieu lenh']
results = []
for i in range(7):
  try: 
    add_and_remove_bboxes(results, f'./results/comp4_det_test_{label[i]} .txt')
  except FileNotFoundError:
    add_and_remove_bboxes(results, f'./results/comp4_det_test_{label[i]}.txt')

# 5. Save data to json file 
with open('submission.json', 'w') as outfile:
    json.dump(results, outfile, indent=4, separators=(", ", ": "))
```

```
import cv2 
import matplotlib.pyplot as plt

def result_get_bbox(filepath):
  img = cv2.imread(filepath)
  img_name = filepath.split('/')[-1]
  idx_img = int(img_name.split('.')[0])
  bboxs = ridx2bbox[idx_img]
  
  for bbox in bboxs:
    top_left = (int(bbox[0]), int(bbox[1]))
    bottom_right = (int(bbox[2]), int(bbox[3]))

    color, thickness = (0, 255, 0), 1
    font = cv2.FONT_HERSHEY_SIMPLEX   
    org, fontScale = (50, 50), 1 
    rcolor = (0, 0, 255) 
    
    img = cv2.rectangle(img, top_left, bottom_right, color, thickness) 
    img = cv2.putText(img, f'{bbox[4]}', top_left, font, fontScale, rcolor, thickness, cv2.LINE_AA) 

  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plt.imshow(img)
```

```
with open('submission.json', 'r') as f:
  result_box = json.load(f)
```

```
def samebox(last_el, arr):
  for idx, i in enumerate(arr):
    if (abs(i[0] - last_el[0]) < 4) + (abs(i[1] - last_el[1]) < 4) + (abs(i[2]- last_el[2]) < 4) + (abs(i[3] - last_el[3]) < 4) >=3:
      if last_el[5] > i[5]:
        arr[idx] = last_el
      return arr 
  arr.append(last_el)
  return arr
```

```
ridx2bbox = {}
ridxs = []
for x in result_box:
  if x['image_id'] not in ridx2bbox:
    ridx2bbox[x['image_id']] = []
  bbox = x['bbox'] 
  ridx2bbox[x['image_id']].append([bbox[0], bbox[1], bbox[0] + bbox[2], bbox[1] + bbox[3], x['category_id'], x['score']])
  if len(ridx2bbox[x['image_id']]) > 1:
    ridx2bbox[x['image_id']] = samebox(ridx2bbox[x['image_id']][-1], ridx2bbox[x['image_id']][:-1])
```
